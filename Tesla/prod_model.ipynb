```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Load the dataset
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# Get past 365 days
end_date = datetime.today()
start_date = end_date - timedelta(days=365)

# Download TSLA data with clear format
df_tsla_raw = yf.download('TSLA', start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'), progress=False, auto_adjust=False)

# Reset index to bring 'Date' into a column
df_tsla = df_tsla_raw.reset_index()

# Rename columns to match required format
df_tsla.columns = ['Date', 'TSLA_Open', 'TSLA_High', 'TSLA_Low', 'TSLA_Close', 'TSLA_Adj_Close', 'TSLA_Volume']

# Drop Adjusted Close if not needed
df_tsla.drop(columns=['TSLA_Adj_Close'], inplace=True)

# Confirm final columns
print(df_tsla.head())
print(df_tsla.columns)

```

```python
plt.figure(figsize=(15,6))
df_tsla['TSLA_Open'].plot()
df_tsla['TSLA_Close'].plot()
plt.xlabel(None)
plt.ylabel(None)
plt.title('Opening & Closing Price of TESLA')
plt.legend(['Open Price', 'Close Price'])
plt.tight_layout()
plt.show()
```
